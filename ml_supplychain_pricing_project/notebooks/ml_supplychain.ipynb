```python
import boto3
import pandas as pd
from io import StringIO

# Make sure your AWS credentials are set as environment variables
# export AWS_ACCESS_KEY_ID=your_key
# export AWS_SECRET_ACCESS_KEY=your_secret

# Initialize S3 client
s3 = boto3.client('s3')

bucket_name = "ml-supplychain-project"  # your bucket name
file_key = "SCMS_Delivery_History_Dataset.csv"                   # CSV file key in S3

# Download file from S3
try:
    response = s3.get_object(Bucket=bucket_name, Key=file_key)
    csv_content = response['Body'].read().decode('utf-8')

    # Load CSV into pandas DataFrame
    df = pd.read_csv(StringIO(csv_content))
    print("Data loaded successfully! Preview:")
    display(df.head())

except Exception as e:
    print("Error loading data from S3:", e)
```
